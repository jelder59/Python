```python
# Função para calcular o aumento baseado nas faixas de salário
def calcular_reajuste(salario):

    # 1. Salário MAIOR que 1500
    if salario > 1500.00:
        aumento = 25.00

    # 2. De 750 inclusive (>=) até 1500 inclusive (<=)
    elif salario >= 750.00 and salario <= 1500.00:
        aumento = 50.00

    # 3. De 450 inclusive (>=) a 750.
    elif salario >= 450.00 and salario < 750.00:
        aumento = 75.00

    # 4. Menor que 450
    elif salario < 450.00:
        aumento = 100.00

    # Caso a pessoa digite um valor negativo!
    else:
        aumento = 0.00
        print("Aviso: Salário inválido.")

    # Calculamos o novo valor
    novo_salario = salario + aumento

    # Mostramos o resultado
    print("---------------------------------")
    print("")
    print(f"Salário Atual: R$ {salario:.2f}")
    print(f"Aumento Fixo:  R$ {aumento:.2f}")
    print(f"Novo Salário:  R$ {novo_salario:.2f}")
    print("")
    print("---------------------------------\n")

print("======== REAJUSTE SALARIAL =========")
print("")
print("Para encerrar digite 0\n")

while True:
    # Coletamos o salário. Usamos float() porque é dinheiro!
    salario_digitado = float(input("Digite o salário atual: R$ "))

    # Condição rápida para sair do 'while True' (se digitar Salário 0 ou negativo)
    if salario_digitado <= 0:
        print("\nPrograma encerrado!")
        break

    # Chama a função principal com o salário capturado
    calcular_reajuste(salario_digitado)

```
