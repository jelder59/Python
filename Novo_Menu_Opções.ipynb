```python
while True:
    print("\nMenu:")
    print("1 Imposto")
    print("2 Novo salário")
    print("3 Classificação")
    print("4 Sair")

    opcao = input("\nDigite a opção desejada: ")

    # ========================================================
    # OPÇÃO 1: IMPOSTO
    # ========================================================
    if opcao == '1':
        salario = float(input("Digite o salário do funcionário: R$ "))

        # Regras: < 500 (5%); de 500 a 850 (10%); acima de 850 (15%)
        # Variável 'percentual_usado' para mostrar na tela

        if salario < 500.00:
            percentual_usado = 5
            imposto = salario * 0.05
        elif salario >= 500.00 and salario <= 850.00:
            percentual_usado = 10
            imposto = salario * 0.10
        else:
            percentual_usado = 15
            imposto = salario * 0.15

        print(f"-> Percentual de imposto aplicado: {percentual_usado}%")
        print(f"-> Valor do imposto: R$ {imposto:.2f}")

    # ========================================================
    # OPÇÃO 2: NOVO SALÁRIO
    # ========================================================
    elif opcao == '2':
        salario = float(input("Digite o salário do funcionário: R$ "))

        # Regras: > 1500 (+25); 750 a 1500 (+50); 450 a 750 (+75); < 450 (+100)
        if salario > 1500.00:
            aumento = 25.00
        elif salario >= 750.00 and salario <= 1500.00:
            aumento = 50.00
        elif salario >= 450.00 and salario < 750.00:
            aumento = 75.00
        else: # menor que 450
            aumento = 100.00

        novo_salario = salario + aumento
        print(f"-> Valor do aumento: R$ {aumento:.2f}")
        print(f"-> Novo salário: R$ {novo_salario:.2f}")

    # ========================================================
    # OPÇÃO 3: CLASSIFICAÇÃO
    # ========================================================
    elif opcao == '3':
        salario = float(input("Digite o salário do funcionário: R$ "))

        # Regras: <= 700 ("Mal Remunerado"); Maior que 700 ("Bem remunerado")
        if salario <= 700.00:
            print("-> Classificação: Mal Remunerado")
        else:
            print("-> Classificação: Bem remunerado")

    # ========================================================
    # OPÇÃO DE SAÍDA E TRATAMENTO DE ERRO
    # ========================================================
    elif opcao == '4':
        print("Saindo do programa...")
        break

    else:
        # Se digitar qualquer coisa fora do esperado
        print("Opção inválida!")

```
