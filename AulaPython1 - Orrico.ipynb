```python
nome  = input("Digite o nome do usuário: ")
salario = float(input('Digite o salário do usuário: R$'))
print(f'O salário do (a) {nome} é de R$ {salario}!!!')
if salario >= 1000:
  print (f'O(a) {nome} ganha bem!!!')
else:
  print(f'O(a) {nome} ganha pouquinho!!!')
```

```python
nome = input("Digite seu nome: ")
n1 = int(input("Digite a primeira nota: "))
n2 = int(input("Digite a segunda nota: "))
n3 = int(input("Digite a terceira nota: "))
n4 = int(input("Digite a quarta nota: "))
media = (n1+n2+n3+n4)/4
if media >=7:
  print(f'{nome} sua média é {media} e você foi aprovado!')
elif media <5:
  print(f'[nome] sua média é {media} e você foi reprovado!')
else:
  print(f'{nome} sua média é {media} e você está de recuperação!')
```
