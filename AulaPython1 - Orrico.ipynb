Match...case

```python
time = input('Entre com o nome do seu time de futebol: ')
match time:
  case 'Flamengo '| 'Fluminense' | 'Vasco' | 'Botafogo':
    print('Este é um time carioca!!!')
  case 'São Paulo' | 'Santos' | 'Palmeiras'| 'Corinthians' | 'Juventus':
    print('Este é um time paulista!!!')
  case _: #outro caso
    print('Este é um time de outro estado!!!')
```

```python
valor = float(input('Digite o valor da mercadoria:'))
op = int(input('Digite 1 para parcelar ou 2 para pagamento a vista: '))
match op:
  case 1:
    if valor > 1000:
      parcelas = valor/5
      print(f'Com esse valor você pode parcelar em 5 vezes de R$: {parcelas: .2f}')
    else:
      parcelas = valor/3
      print(f'Com esse valor você pode parcelar em 3 vezes de R$: {parcelas: .2f}')
  case 2:
    avista = (valor *0.9)
    print(f'O valor a vista é: {avista: .2f}')
  case _:
    print ('Opção Inválida!')
```

Repetição... while

```python
cont = 0
n = int(input('Digite um número: '))
while cont <10:
  cont += 1
  m = cont * n
  print(f'{n} X {cont} = {m}')
print(f'Esta é a tabuada do {n}')
```

```python
print('|' '-' * 25,
      '|') # O 50 multiplica o caractere
print('| ------ Sistema de conversão de moedas local ------')

qtd = int(input('Quantas conversões deseja fazer no momento?'))
cont = 0

while cont < qtd:
  cont += 1 # incremento
  reais = float (input('Qual o valor em Reais deseja converter: R$ '))
  dolar = float (input('Qual o valor do dolar hoje: ? '))
  valor = reais / dolar

  print(f'R$ {reais} em dolar é: US$ {valor: .2f}')
print('| ------ O SCML Agradece sua visita!!! ------|')
print('|', '-' * 50, '|')
```

```python
qtd = int(input('Quantos produtos deseja comprar: '))

while qtd <=0:
  qtd = int(input('Digite um valor válido'))

total = 0

for i in range(1, qtd+1):
  valor = float(input(f'Digite o preço do produto {i}: '))

  while valor <=0:
    valor = float(input('Valor inválido. Digite novamente: '))
  total += valor

print (f'\n Total da compra: R$ {total: .2f}')

desconto = 0

if total > 100:
  desconto = total * 0.10
  total -= desconto

  print(f'Desconto aplicado: R$ {desconto: .2f}')

else:
  print('Sem desconto')

print(f'Valor após desconto: R$ {total: .2f}')

print('\nForma de pagamento:')
print('1 - À vista')
print('2 - Parcelado')

op = int(input('Escolha: '))

if op == 1:
  desc_pag = total * 0.05
  total -= desc_pag
  print(f'Desconto a vista: R$ {desc_pag:.2f}')

elif op == 2:
  print('Pagamento parcelado.')

else:
  print('Opção inválida')

print(f'\nValor final: R$ {total:.2f}')
```
